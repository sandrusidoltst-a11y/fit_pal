# Food Lookup Tool Evaluation

This notebook evaluates the `search_food` tool using examples provided in `noetbooks/querys_exmpels.md`.
The goal is to check the quality of search results for specific keywords extracted from Hebrew inputs.

```python

import sys
import os

# Add project root to path
sys.path.append(os.path.abspath('..'))

from src.tools.food_lookup import search_food
import pandas as pd
```

```python
test_keywords = [
    "Rice", 
    "White rice",
    "White bread", 
    "Tuna", 
    "Tuna in water",
    "Chicken thigh", 
    "Pasta"
]


def get_query_rankings(query):
    print(f"\n--- Results for: '{query}' ---")
    search_results = search_food.invoke(query)
    
    if not search_results:
        print("No results found.")
        return pd.DataFrame()
    
    data = [
        {
            "Rank": i + 1,
            "Query": query,
            "Result Name": res.get('name', 'N/A'),
            "Food ID": res.get('id', 'N/A')
        }
        for i, res in enumerate(search_results)
    ]
    
    df = pd.DataFrame(data)
    print(df.to_string(index=False))
    return df

# Evaluate each keyword and combine into a single DataFrame
evaluation_results = [get_query_rankings(kw) for kw in test_keywords]
full_results_df = pd.concat(evaluation_results, ignore_index=True)
full_results_df
```
